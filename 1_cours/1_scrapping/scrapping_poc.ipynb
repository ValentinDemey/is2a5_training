## Exemple from 1st link

```python
from bs4 import BeautifulSoup 
from lxml import etree 
import requests 


URL = "https://en.wikipedia.org/wiki/Nike,_Inc."

HEADERS = ({'User-Agent': 
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language': 
            'en-US, en;q=0.5'}) 

webpage = requests.get(URL, headers=HEADERS) 
soup = BeautifulSoup(webpage.content, "html.parser") 
dom = etree.HTML(str(soup)) 
print(dom.xpath('//*[@id="firstHeading"]/span')[0].text)
```

## CRAWLER

### Page 1 test

Ouvrir l'url : https://books.toscrape.com/catalogue/page-1.html

```python
from bs4 import BeautifulSoup 
from lxml import etree 
import requests 


URL = "https://books.toscrape.com/catalogue/page-1.html"

HEADERS = ({'User-Agent': 
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language': 
            'en-US, en;q=0.5'}) 

webpage = requests.get(URL, headers=HEADERS) 
soup = BeautifulSoup(webpage.content, "html.parser") 
dom = etree.HTML(str(soup)) 
```

### Get one book Url from current page

```python
e=dom.xpath('//*[@id="default"]/div/div/div/div/section/div[2]/ol/li[1]/article/h3/a')
for el in e:
    print(el.get('href'))
```

### Get all book Urls from current page

```python
e=dom.xpath('//*[@id="default"]/div/div/div/div/section/div[2]/ol/li[*]/article/h3/a')
print(len(e))
for el in e:
    print(el.get('href'))
```

## SCRAPPER

Ouvrir l'url : https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html


```python
from bs4 import BeautifulSoup 
from lxml import etree 
import requests 


URL = "https://books.toscrape.com/" + "catalogue/a-light-in-the-attic_1000/index.html"

HEADERS = ({'User-Agent': 
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language': 
            'en-US, en;q=0.5'}) 

webpage = requests.get(URL, headers=HEADERS) 
soup = BeautifulSoup(webpage.content, "html.parser") 
dom = etree.HTML(str(soup)) 
```

```python
e=dom.xpath('//table//tr[2]/td')
print(e[0])
for el in e:
    print(el.text)
```

### Parser how to get xpath using nav

```python
xpath='//article/p'
e=dom.xpath(xpath)
print(e[0])
for el in e:
    print(el.text)
```

```python
xpath='//tr[3]/td'
e=dom.xpath(xpath)
print(e[0])
for el in e:
    print(el.text)
```
